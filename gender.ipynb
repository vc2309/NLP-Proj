```python
import nltk
import random
from nltk.corpus import names
```

```python
names.fileids()
```

```python
names.words('female.txt')
```

```python
f={n:'female' for n in names.words('female.txt')}
```

```python
m={n:'male' for n in names.words('male.txt')}
```

```python
f
```

```python
names=[(n) for (n) in f]
```

```python
n1=[(n,'female') for n in f]
n2=[(n,'male') for n in m]
```

```python
all=n1+n2
```

```python
all
```

```python
an=all
```

```python
random.shuffle(an)
```

```python
an
```

```python
def features(x):
    v=['a','e','i','o','u']
    numv=0
    for char in x:
        if char in v:
            numv=numv+1
    return {'pair': x[-2:],
           'numv':numv}
all_feats=[(features(n[0]),n[1]) for n in an]
```

```python
all_feats
```

```python
train=all_feats[:round(0.7*len(all_feats))]
test=all_feats[len(train):]
```

```python
test
```

```python
clf=nltk.NaiveBayesClassifier.train(train)
```

```python
clf.classify(features('Vishnu'))
```

```python
clf.classify(features('Devki'))
```

```python
nltk.classify.accuracy(clf, test)
```
